```python
data  = data[['Name','Location','Price','Kms_driven','Fuel_type','Year','Company']]
```

```python
data.info()
```

### to do 
- sort the data by removing the unwanted colums 
- convert Kms_driven into int and removing the , between the numbers
- convert Price into int and remove the , between the numbers
- reduce the size of the name by getting first three words as name of the car
- remove the year from the name 

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data['Kms_driven']
```

```python
data['Kms_driven'] = data['Kms_driven'].str.split(" ").str.get(0).str.replace(',','').astype(int)
```

```python
data.head()
```

-----------------------------------------------------------------------------------------------------------------

```python
data['Price'] = data['Price'].str.split('₹').str.get(1).str.replace(',','')
```

```python
data.dropna(inplace = True)    # this removes all the Nan values created due to split 
```

```python
data['Price'] = data['Price'].astype(int)
```

```python
data.head()
```

```python
data.info()
```

```python
data['Name'] = data['Name'].str.split(' ').str.slice(0,3)
```

```python
data['Name'] = data['Name'].str.join(' ')
```

```python
data
```

```python
data =  data.reset_index(drop = True)
```

```python
data
```

```python
data.describe()
```

```python
data = data[data['Price']<6e6].reset_index(drop = True)
```

```python
data
```

```python
data.to_csv('car_DataFrame.csv')
```
